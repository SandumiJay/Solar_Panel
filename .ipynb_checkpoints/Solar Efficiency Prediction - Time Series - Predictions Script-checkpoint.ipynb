```python
def add_date_time(_df):
    "Returns a DF with two new cols : the time and hour of the day"
    t = pd.date_range(start='1/1/1986', periods=_df.shape[0], freq = 'H')
    t = pd.DataFrame(t)
    _df = pd.concat([_df, t], axis=1)
    _df.rename(columns={ _df.columns[-1]: "time" }, inplace = True)
    return _df
```

Effiency prediction date and location method 01

```python
def time_series_by_Location(location):
    y = train[location].resample('M').mean()
    p = d = q = range(0, 2)
    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
    aic=[]
    l_param=[]
    l_param_seasonal=[]
    temp=0
    lowest=10000000000000000000


    for param in pdq:
        for param_seasonal in seasonal_pdq:
            try:
                mod = sm.tsa.statespace.SARIMAX(y,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
                results = mod.fit()
#                 print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
                aic.append(results.aic)
                temp=results.aic
                if(temp<lowest):
                    lowest=temp
                    l_param=param
                    l_param_seasonal=param_seasonal
            
            except:
                continue
    
    mod = sm.tsa.statespace.SARIMAX(y,order=l_param,seasonal_order=l_param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
    model_fit= mod.fit()
    print(model_fit.summary())
    model_fit.save(location +'.pkl')
    
```

```python
from os.path import exists 
def get_time_series_Prediction(location,Start_date,End_date,location_exists):
    model_file_name =location+".pkl"
    location_list=set(data.columns.tolist())
    
    if(location in location_list):
        location_exists=True
        
    if(exists(model_file_name)):
        model_loaded = ARIMAResults.load(model_file_name)
        return model_loaded.predict(Start_date,End_date)
    else:
        if(location_exists==True):
            time_series_by_Location(location)
            model_loaded = ARIMAResults.load(model_file_name)
            return model_loaded.predict(Start_date,End_date)
        else:
            model_loaded = ARIMAResults.load('TS_model_for_Efficency_Year_month.pkl')
            return model_loaded.predict(Start_date,End_date)
```

```python
def get_yearly_avg_solar_effiency(location,start_date,end_date,location_exists):
    output={}
    result=pd.DataFrame()
    year=pd.DataFrame()
    Average_Effiency=pd.DataFrame()
    pred_data = get_time_series_Prediction(location,start_date,end_date,location_exists)
    df = pred_data.to_frame().reset_index()
    df=df.set_index('index')
    avg_eff= df['predicted_mean'].resample('Y').mean()*100
    avg_eff=avg_eff.to_frame().reset_index()
    result['Year']=avg_eff['index'].dt.year
    result['Average Effiecy']=avg_eff['predicted_mean']
    return result
```
