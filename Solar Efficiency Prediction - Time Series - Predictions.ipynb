```python
def add_date_time(_df):
    "Returns a DF with two new cols : the time and hour of the day"
    t = pd.date_range(start='1/1/1986', periods=_df.shape[0], freq = 'H')
    t = pd.DataFrame(t)
    _df = pd.concat([_df, t], axis=1)
    _df.rename(columns={ _df.columns[-1]: "time" }, inplace = True)
    return _df
```

Effiency prediction date and location method 01

```python
def time_series_by_Location(location):
    y = train[location].resample('M').mean()
    p = d = q = range(0, 2)
    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
    aic=[]
    l_param=[]
    l_param_seasonal=[]
    temp=0
    lowest=10000000000000000000


    for param in pdq:
        for param_seasonal in seasonal_pdq:
            try:
                mod = sm.tsa.statespace.SARIMAX(y,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
                results = mod.fit()
#                 print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
                aic.append(results.aic)
                temp=results.aic
                if(temp<lowest):
                    lowest=temp
                    l_param=param
                    l_param_seasonal=param_seasonal
            
            except:
                continue
    
    mod = sm.tsa.statespace.SARIMAX(y,order=l_param,seasonal_order=l_param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
    model_fit= mod.fit()
    print(model_fit.summary())
    model_fit.save(location +'.pkl')
    
```

```python
time_series_by_Location('France')
```

```python
from os.path import exists 
def get_time_series_Prediction(location,Start_date,End_date):
    model_file_name =location+".pkl"
    if(exists(model_file_name)):
        model_loaded = ARIMAResults.load(model_file_name)
        return model_loaded.predict(Start_date,End_date)
    else:
        time_series_by_Location(location)
        model_loaded = ARIMAResults.load(model_file_name)
        return model_loaded.predict(Start_date,End_date)
```
